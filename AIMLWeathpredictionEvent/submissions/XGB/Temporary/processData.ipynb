```python
import pandas as pd
# Data Preprocessing
df = pd.read_csv('./train.csv')
pd.set_option('display.max_rows',200)
df.head(20)
```

```python
dateList = pd.date_range(start='1951-01-01', end='2009-12-31').tolist()
df['Date'] = dateList

dates = []
months = []
years = []
for x in df['Date'] :
    temp = str(x).split("-")
    years.append(temp[0])
    months.append(temp[1])
    dates.append(temp[2][:2])
df['Full date'] = df['Date']
df['Year'] = years 
df['Month'] = months
df['Date'] = dates
df
```

```python
df.to_csv('processedData.csv', index=False) #make prototypes in knime
```
